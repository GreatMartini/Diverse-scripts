## Introduction
### Composition des nébuleuses
Les nébuleuses sont des nuages interstellaires composés de poussières et de gaz. On peut en distinguer deux types, les nébuleuses diffuses qui emettent ou réfléchissent de la lumière et les nébuleuses obscures l'absorbent. Pour ce projet on se limitera à traiter les nébuleuses diffuses en émission. Dans le TP de radioastronomie on a remarqué que la matière interstellaire était principalement composée d'hydrogène. De plus, la majorité des nébuleuses en émission absorben la radiation stellaire pour après la réémètre. Dans ce cadre, on peut traduire une nébuleuse par un ensemble de gaz ionisé par radiation stellaire qui après est réémise. Désormais, on peut assumer que ce type de nébuleuses émettra principalement de la lumière visible correspondante à la transition Balmer $\alpha$ de l'hydrogène, c'est à dire une transition de $n=3$ à $n=2$ qui, correspond à une longueur d'onde de $656,3nm$. De plus, la nébuleuse étant composée d'autres éléments elle pourra émettre aussi de la lumière à la longueur d'onde correspondante à l'oxygène ionisé deux fois et au silicium ionisé une fois.
### Structure générale des appareils d'observation
Le telescope réflecteur est l'outil principal utilisé pour l'imagerie astrophysique. Ce dernier peut se décomposer en une partie optique et une partie électronique. La première est constituée d'un objectif qui collecte le rayonnement et d'un arrangement de mirroirs qui permettent de focaliser et de renvoyer les rayons à la partie électronique. La partie électronique est un capteur CCD (Charge-coupled device) qui est composé de semi-conducteurs intrinscèques ou dopés de type p. Quand le CCD reçoit des photons il transforme l'intensité lumineuse en courant électrique, néanmoins il présente un bruit dû au "courant d'obscurité", courant résiduel qui est généré par un mouvement aléatoire des électrons dans le CCD en abscence d'excitations lumineuses.
### Principe de l'analyse des images
Les images sont alors prises par le telescope et enregistrées en format numérique à travers le CCD. Cependant ces images contiennent aussi du bruit, non seulement de courant noir mais aussi de bias et de flat. Le bruit flat est associé à des étérogénéités de la réponse des pixels à une intensité homogène à une longeur d'onde donnée. D'autre part le bias représente les imperfections éléctroniques sans temps d'exposition et sans excitations lumineuses dans le CCD.  Nous allons donc, dans ce Tp, retirer toutes ses différentes sources de bruit afin d'avoir les images les plus nettes et fidèles possibles à la réalité.  

Avant de commencer à expliquer comment nous allons le faire nous aimerions revenir sur ce que représentent les différentes images que nous allons Utiliser pour corriger. Pour commencer nous avons notre image brute prise par le capteur. Nous avons ensuite le dark, cette image est créée en mettant le cache sur le détecteur en ayant un temps d'explosion; cette image nous permet d'avoir l'erreur sur le détecteur. Viens ensuite le bias, cette image est prise toujours en ayant le capteur mais sans temps d'exposition maintenant, cette image nous permet de voir les erreurs dues aux électrons et non au détecteur. Finalement nous avons le flat qui est pris pour chaque filtre sous une exposition à une intensité uniforme. En théorie, les prises de Flat contiennent du bruit de Dark et de Bias et, les prises de Dark du bruit associé au Bias. Dance cadre, et particulièrement les prises dark utilisées dans ce TP contiennent déjà le bruit du bias, alors, afin d'obtenir notre image finale on fait faire à notre code l'opération suivante sur les images prises :  

\begin{equation}
Image\,finale= \frac{image\,brut - dark}{flat-dark}
\end{equation}

### Question 4: Courant d'obscurité
Le courant d'obscurité est un mouvement aléatoire des électrons à l'intérieur de semi-conducteurs et dans le cas présent, à l'intérieur d'un CCD. Ce courant est un phénomène purement thermique qui ne peut pas être prédit, par conséquent, ce n'est pas un signal reproductible par changement de prise de vue.  
Suite à cela, pour acquérir le courant d'obscurité il est nécessaire de le faire avec un obturateur fermé de façon à que le CCD ne soit pas exposé à des photons. De plus, comme il s'agit d'un phénomène thermique, l'acquisition doit se faire à la même température que celle de la prise de vues et doit avoir le même temps d'exposition pour avoir une moyenne correcte du patron de comportement du bruit.
### Question 5: Signal et traitement
L'image prise sera une superposition de l'image intrinsèque de l'objet et du courant d'obscurité. Alors, pour obtenir l'image désirée on doit retirer le courant d'obscurité du signal et réduire ce total par la magnification due à la plage de lumière uniforme. Cela est valable car la plage de lumière uniforme est une fonction qui multiplie le signal digital permettant de corriger les défauts de gain dans les pixels, mathématiquement nous pouvons traduire le tout par:
\begin{equation}
IP+C=O,
\end{equation}
où **I** est le signal de l'image recherchée, **C** le courant d'obscurité, **P** la plage de lumière uniforme et **O** le signal non traité, ce qui veut dire que:
\begin{equation}
I=(O-C)/P,
\end{equation}

#### Traitement de l'image:
On a commencé par déclarer les dépendances python:

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import os
import glob
from astropy.utils.data import download_file
from astropy.io import fits
from matplotlib.colors import LogNorm
import astroalign as aa
import aplpy
from astroscrappy import detect_cosmics as lacosmic
```

Pour continuer, on a déclaré les localisations des fichiers de données et de sortie

```python
data_dir = "/Users/Bach/Desktop/TP/Observation/" #Fichier de données
output_dirH = "/Users/Bach/Desktop/TP/Observation/outh/" #Fichier de sortie pour le filtre Halpha
output_dirO = "/Users/Bach/Desktop/TP/Observation/outo/" #Fichier de sortie pour le filtre OIII
output_dirR = "/Users/Bach/Desktop/TP/Observation/outr/" #Fichier de sortie pour le filtre Red 
output_noise_dir = "/Users/Bach/Desktop/TP/Observation/noise/" #Fichier de sortie pour les masters
```

Ainsi, on à défini deux fonctions, vmin et vmax qui retournent les valeurs maximales et minimales des données basées sur la médiane et $\pm\sigma$, où $\sigma$ est la déviation standard des données.

```python
def vmax(data):
    return np.median(data)+np.std(data)
def vmin(data):
    return np.median(data)-np.std(data)
```

Après on a créé une fonction pour créer le master bias:

```python
def creation_master_bias():
    bias_list = glob.glob(data_dir+"Bias/*fits") #On obtient la liste des Bias
    bias_concat = [] #On crée une liste vide
    
    for bias in bias_list: #On itère sur la liste
        bias_concat.append(fits.getdata(bias)) #On extraie les données et on les ranges dans la liste de données
    final_bias = np.median(bias_concat, axis=0) #On crée le master en prenant la mediane pour chaque pixel dans la liste
    
    outfile = output_noise_dir+'master_bias.fits' 
    hdu = fits.PrimaryHDU(final_bias)
    hdu.writeto(outfile, overwrite=True) #On sauvegarde le Master Flat
```

On répète les fonctions pour les masters flats pour chaque filtre

```python
def creation_master_flatH():
    flath_list = glob.glob(data_dir+"Flat/H_Alpha/*fits")
    flath_concat = []
    for flath in flath_list:
        flath_concat.append(fits.getdata(flath))
    final_flath = np.median(flath_concat, axis=0)

    outfile = output_noise_dir+'master_flath.fits'
    hdu = fits.PrimaryHDU(final_flath)
    hdu.writeto(outfile, overwrite=True)
def creation_master_flatO():
    flato_list = glob.glob(data_dir+"Flat/Blue/*fits")
    flato_concat = []
    for flato in flato_list:
        flato_concat.append(fits.getdata(flato))
    final_flato = np.median(flato_concat, axis=0)

    outfile = output_noise_dir+'master_flato.fits'
    hdu = fits.PrimaryHDU(final_flato)
    hdu.writeto(outfile, overwrite=True)
def creation_master_flatR():
    flatr_list = glob.glob(data_dir+"Flat/Red/*fits")
    flatr_concat = []
    for flatr in flatr_list:
        flatr_concat.append(fits.getdata(flatr))
    final_flatr = np.median(flatr_concat, axis=0)

    outfile = output_noise_dir+'master_flatr.fits'
    hdu = fits.PrimaryHDU(final_flatr)
    hdu.writeto(outfile, overwrite=True)
```

Pour continuer, on à créé les fonctions relatives au traitement des images pour chaque filtre. Dans ce cas on ne doit pas utiliser le Master Bias car il est déjà compris dans le Master Dark (déjà fait).

```python
def traitement_H():
    master_flath = os.path.join(data_dir,"noise/master_flath.fits") #On déclare le fichier Master Flat relatif au filtre
    master_bias = os.path.join(data_dir,"noise/master_bias.fits") #On déclare le fichier Master Bias
    master_dark = os.path.join(data_dir,"/Users/Bach/Desktop/TP/Observation/MASTERS/old/master_dark_10.fits")#On déclare Master dark
    flath = fits.getdata(master_flath) #On obtient les données des fichiers Master
    bias = fits.getdata(master_bias)
    dark = fits.getdata(master_dark)

    image_list = glob.glob(data_dir+'M42_2307/H_Alpha/*fits') #On obtient la liste des images
    image_concat=[] #On déclare une liste vide
    for image in image_list: #On itère sur la liste des images
        image_data=fits.getdata(image) #On extraie les données
        processed = (image_data-dark)/(flath-dark) #On traite, le flat est composé du dark donc il faut le soustraire
        image_concat.append(processed) #On range dans la liste
        filename,file_extension=os.path.splitext(image) #On sauvegarde les images traitées
        outfile=output_dirH+os.path.basename(filename)+'_pr.fits'
        hdu = fits.PrimaryHDU(processed)
        hdu.writeto(outfile, overwrite=True)
def traitement_O():#On répète pour l'oxygène
    master_flato = os.path.join(data_dir,"noise/master_flato.fits")
    master_bias = os.path.join(data_dir,"noise/master_bias.fits")
    master_dark = os.path.join(data_dir,"/Users/Bach/Desktop/TP/Observation/MASTERS/old/master_dark_10.fits")
    flato = fits.getdata(master_flato)
    dark = fits.getdata(master_dark)
    image_list = glob.glob(data_dir+'M42_2307/OIII/*fits')
    image_concat=[]
    for image in image_list:
        image_data=fits.getdata(image)
        processed = (image_data-dark)/(flato-dark)
        image_concat.append(processed)
        filename,file_extension=os.path.splitext(image)
        outfile=output_dirO+os.path.basename(filename)+'_pr.fits'
        hdu = fits.PrimaryHDU(processed)
        hdu.writeto(outfile, overwrite=True)
def traitement_R():#On répète pour le rouge
    master_flatr = os.path.join(data_dir,"noise/master_flatr.fits")
    master_bias = os.path.join(data_dir,"noise/master_bias.fits")
    master_dark = os.path.join(data_dir,"/Users/Bach/Desktop/TP/Observation/MASTERS/old/master_dark_10.fits")
    flatr = fits.getdata(master_flatr)
    dark = fits.getdata(master_dark)
    image_list = glob.glob(data_dir+'M42_2307/Red/*fits')
    image_concat=[]
    for image in image_list:
        image_data=fits.getdata(image)
        processed = (image_data-dark)/(flatr-dark)
        image_concat.append(processed)
        filename,file_extension=os.path.splitext(image)
        outfile=output_dirR+os.path.basename(filename)+'_pr.fits'
        hdu = fits.PrimaryHDU(processed)
        hdu.writeto(outfile, overwrite=True)
```

Après avoir créé les fonctions nous les avons appliquées pour produire les résultats.

```python
""" Création de fichiers masters: """
""" Création de Master Bias"""
creation_master_bias()
""" Création de Master Flat pour H_alpha """
creation_master_flatH()
""" Création de Master Flat pour OIII """
creation_master_flatO()
""" Création de Master Flat pour Red """
creation_master_flatR()

""" Traitement de l'image pour les trois filtres"""
""" H_Alpha """
traitement_H()
""" OIII """
traitement_O()
""" Red """
traitement_R()
```

De cette façon on à fini le pré-traitement des images.
